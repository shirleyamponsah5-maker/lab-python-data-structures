```python
products = ["t-shirt", "mug" , "hat" , "book" , "keychain"]
products
```

```python
inventory = {}
```

```python
for product in products:
    quantity = int(input(f"Enter the quantity for {product}: "))
    inventory[product] = quantity
```

```python
customer_orders = set()
```

```python
for k in range(3):
    order = input("Enter the name of a productyou want to order: ")
    customer_orders.add(order)
```

```python
print("\nProducts in Customer Orders:", customer_orders)
```

```python
total_products_ordered = len(customer_orders)
percentage_products_ordered = (total_products_ordered / len(products))* 100
order_status = (total_products_ordered, percentage_products_ordered)
```

```python
print("\nOrder Statistics:")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage Products Ordered: {order_status[1]}%")
```

```python
for product in customer_orders:
    if product in inventory:
        inventory[product] -= 1
```

```python
print("\nUpdated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
